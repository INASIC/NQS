```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def psi_sampler(n, a, x):  # a = boundary condition
    if x <= 0 or x >= a:
        return 0
    else:
        return np.sqrt((2/a)) * np.sin(n * np.pi * x / a)
```

```python
a = 100  # Boundary condition (size of well)
padding = 10
Xs = range(-padding,a+padding)
```

```python
# Sample from true functions
n=1
psi_list = []
for x in Xs:
    psi = psi_sampler(n, a, x)
    psi_list.append(psi)
```

```python
# Visualize multiple wavefunctions overlaid
fig=plt.figure(figsize=(10, 5))
plt.rcParams.update({'font.size': 14})
n=1
plt.plot(Xs, psi_list)
# ax.set_yticklabels([])
# plt.title('Wavefunction Particle in a Box')
plt.xlabel('x')
plt.ylabel('$\Psi(x)$')
cur_axes = plt.gca()
# cur_axes.axes.get_xaxis().set_ticklabels([])
cur_axes.axes.get_yaxis().set_ticklabels([])
# cur_axes.axes.get_xaxis().set_ticks([])
cur_axes.axes.get_yaxis().set_ticks([])
plt.xlim(-10,110)
# plt.ylim(0,0.2)
plt.savefig('plot', dpi = 300, bbox_inches='tight')
plt.show()
```

```python
# Nmesh = 1024
Nmesh = 2**4
dx = 120 / Nmesh 
```

```python
x = -10
x_mesh = []
while x <= 110:
    x_mesh.append(x)
    x += dx
```

```python
# Visualize multiple wavefunctions overlaid
fig=plt.figure(figsize=(10, 5))
plt.rcParams.update({'font.size': 16})
n=1
plt.plot(Xs, psi_list)
plt.scatter(x_mesh, np.zeros(len(x_mesh)), c='green', label='mesh-points')#, marker='x')
# ax.set_yticklabels([])
# plt.title('Wavefunction Particle in a Box')
plt.xlabel('x')
plt.ylabel('$\Psi(x)$')
cur_axes = plt.gca()
# cur_axes.axes.get_xaxis().set_ticklabels([])
cur_axes.axes.get_yaxis().set_ticklabels([])
# cur_axes.axes.get_xaxis().set_ticks([])
cur_axes.axes.get_yaxis().set_ticks([])
plt.xlim(-10,110)
# plt.ylim(0,0.2)
plt.axvline(x=0, c='k')#, label='potential barrier')
plt.axvline(x=100, c='k')
plt.legend(loc=1)
plt.savefig('plot', dpi = 300, bbox_inches='tight')
plt.show()
```

## Get the values of psi at each mesh point

```python
psi_mesh = []
for meshpoint in x_mesh:
    psi_mesh.append(psi_sampler(n, a, meshpoint))
```

```python
(markerline, stemlines, baseline) = plt.stem(x_mesh, psi_mesh)#, linefmt='--') #, markerfmt='C0',)
plt.setp(baseline, visible=False)
plt.setp(stemlines, c='C0', linestyle='--')
plt.setp(markerline, marker='o')
plt.show()
```
