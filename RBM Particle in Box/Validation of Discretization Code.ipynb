```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
psi = pd.read_csv('Psi.dat', sep='\s+', header=None)
psi.head()
```

```python
max(psi[0])
```

```python
x = psi[0]
```

```python
max(psi[1])
```

```python
fig=plt.figure(figsize=(10, 5))
plt.rcParams.update({'font.size': 17})
plt.plot(psi[0], psi[1], label='')
plt.ylabel('$\psi (x)$')
plt.xlabel('x')
plt.axvline(x=0, c='k', label='potential barrier')
plt.axvline(x=100, c='k')
plt.legend(loc=1)
plt.xlim(-10,110)
plt.savefig('validation_discretized_psi', dpi = 300, bbox_inches='tight')
plt.show()
```

```python
fig=plt.figure(figsize=(10, 5))
plt.rcParams.update({'font.size': 17})
plt.plot(psi[0], (psi[1] - psi[6]), label='')
plt.ylabel('Relative Error')
plt.xlabel('x')
plt.axvline(x=0, c='k', label='potential barrier')
plt.axvline(x=100, c='k')
plt.legend(loc=1)
plt.xlim(-10,110)
plt.savefig('error_psi_x', dpi = 300, bbox_inches='tight')
plt.show()
```

```python
mean_squared_error = np.mean((psi[1] - psi[6])**2)
mean_squared_error
```
